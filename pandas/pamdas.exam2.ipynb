```python
import pandas as pd

df = pd.DataFrame( 
    { 
     'Name'  : ['Tom', 'Jane', 'Mike', 'Sara', 'John', 'Emma'],
    'Age'   : [20, 22, 19, 21, 23, 20],
    'Score' : [85, 92, 78, 88, 95, 70],
    'Grade' : ['B', 'A', 'C', 'B', 'A', 'C'] 
    }
)

print(df)
print(df.shape)
print(df.describe())
print(df.info())
```

```python
print(df['Score'])
print(df['Name'])
print(df['Age'])
```

```python
print(df.iloc[0])
print(df.iloc[5])
df.index = ['학생1', '학생2', '학생3', '학생4', '학생5', '학생6']
print(df.loc['학생3'])
```

```python
print(df["Score"].max())
print(df["Score"].min())
print(df["Score"].mean())
```

```python
print(df[df["Score"] >= 85])
print(df[df["Grade"] == 'A'])
print(df[df["Score"] >= df["Score"].mean()])
```

```python
print(df.head(3))
print(df.tail(2))
```

```python
# 조건 필터링
# 나이가 30 이상인 사람만 보기
print(df[df["Age"] >= 30])
```
