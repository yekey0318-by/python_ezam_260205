```python
import matplotlib.pyplot as plt

plt.style.use('default')

plt.plot([2,4,6,8])

plt.show()
```

```python
x = [2,4,6,8]
y = [1,3,5,7]
plt.plot(x,y)

plt.show()
```

```python
x = range(11) # 0~10
y = range(5,16) # 5~15

# y :  * :  
# plt.plot(x,y, "mv") -> m :  v :  
plt.plot(x,y, "y*")
plt.show
```

```python
fig = plt.figure() #    ->     
# 2 :  2 (row) 
# 1 :   1 (column) 
# 1 :     .( ) 
fg1 = fig.add_subplot(2,1,1)
fg2 = fig.add_subplot(2,1,2)

fg1.plot([1,2,3],[4,5,6])
fg2.plot([2,4,6], [1,3,7])
plt.show()
```

```python
fig = plt.figure()

fg1 = fig.add_subplot(1,3,1)
fg2 = fig.add_subplot(1,3,3)

fg1.plot([1,2,3],[4,5,6])
fg2.plot([2,4,6], [1,3,7])
plt.show()
```

```python
fig = plt.figure()

fg1 = fig.add_subplot(1,2,1)
fg2 = fig.add_subplot(2,2,2)

fg1.plot([1,2,3],[4,5,6])
fg2.plot([2,4,6], [1,3,7])
plt.show()
```

```python
fig = plt.figure()

a1 = fig.add_subplot(1,2,1)
a2 = fig.add_subplot(1,2,2)

x = range(0,10)
y = range(5,15)
a1.plot(x,y,"r1")
a2.plot(x,y,"bd")
plt.show
```

```python
fig = plt.figure()

#  () 
plt.style.use('ggplot')

# 1 1 , ()  
a = fig.add_subplot(111)

x = [3,4] #    ( )
y = [1,2] #   

labels = ['english', 'math']
colors = ['#3a91bd', '#29ab63']

a.bar(x,y, tick_label=labels, color = colors)
```

```python
fig = plt.figure()

plt.style.use('ggplot')


a = fig.add_subplot(111)

x = [3,4] 
y = [1,2] 

labels = ['english', 'math']
colors = ['#3a91bd', '#29ab63']

a.barh(x,y, tick_label=labels, color = colors)
```

```python
fig = plt.figure()
plt.style.use('ggplot')


a = fig.add_subplot(1, 1, 1)

x = [3, 4]
y = [1, 2]

labels = ['english', 'math']
colors = ['blue', 'red']

a.bar(x, y, tick_label=labels, color=colors)

for i, v in enumerate(y):
    a.text(x[i], v + 0.05, str(v), ha='center') 

plt.show()
```
