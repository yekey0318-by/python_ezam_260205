```python
import numpy as np
arr = np.array([[3,6,9], 
                 [2,4,8],
                 [5,7,1]])

print(arr.ndim)
print(arr.shape)
```

```python
arr = np.array([[3,6,9], 
                 [2,4,8],
                 [5,7,1]])

print("열 방향 평균(axis = 0) : ", arr.mean(axis=0))
print("행 방향 최댓값(axis = 1) : ", arr.max(axis=1))
```

```python
arr = np.array([[3,6,9], 
                 [2,4,8],
                 [5,7,1]])

print(arr.T)
print(arr.T.shape)
```

```python
arr = np.array([[3,6,9], 
                 [2,4,8],
                 [5,7,1]])

print(arr.flatten())
print(arr.flatten().min())
```

```python
arr = np.array([[3,6,9], 
                 [2,4,8],
                 [5,7,1]])

arr2 = arr.reshape(1,9)
print(arr2)
```

```python
arr = np.array([[1.5, 2.5, 3.5, 4.5],
                [5.5, 6.5, 7.5, 8.5]])

print(arr.dtype)
print(arr.itemsize)
```

```python
arr = np.array([[1.5, 2.5, 3.5, 4.5],
                [5.5, 6.5, 7.5, 8.5]])

print(arr.sum(axis = 0))
```

```python
arr = np.array([[1.5, 2.5, 3.5, 4.5],
                [5.5, 6.5, 7.5, 8.5]])

print("전체 평균", arr.mean(axis = 1))
```

```python
arr = np.array([[1.5, 2.5, 3.5, 4.5],
                [5.5, 6.5, 7.5, 8.5]])

arr2 = arr.reshape(4,2)
print(arr2)

print(arr2.T.shape)
```

```python
arr = np.array([[1.5, 2.5, 3.5, 4.5],
                [5.5, 6.5, 7.5, 8.5]])

print(arr.flatten())
print(arr.flatten().shape)
```
